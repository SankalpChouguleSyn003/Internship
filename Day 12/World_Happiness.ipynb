```python
correlation = df['gdp_per_capita'].corr(df['happiness_score']) # Coorelation between GDP and Happiness_Score
```

```python
# Visualizations

plt.figure
plt.scatter(df['gdp_per_capita'], df['happiness_score'])
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('GDP VS Happiness Score (2015)')
plt.show()
```

```python
# Correlation Heatmap

corr = df[['gdp_per_capita', 'happiness_score']].corr()
plt.imshow(corr)
plt.xticks([0,1], ['GDP per capita', 'Happiness score'])
plt.yticks([0,1], ['GDP per capita', 'Happiness score'])
plt.title('Correlation Heatmap')
plt.colorbar()
plt.show()
```
