```python
df['date_added'] = pd.to_datetime(df['date_added'], errors = 'coerce')
```

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['date_added'] = df['date_added'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')
df['durtion'] = df['duration'].fillna('duration')
```

```python
df[['Genre_1', 'Genre_2', 'Genre_3']] = df['listed_in'].str.split(',', expand = True)
```

```python
df['duration(min)'] = df['duration'].str.extract(r'(\d+)').astype(float)

df.loc[df['duration'].str.contains('season', case=False, na = False), 'duration(min)']*= 10 * 25 * 60 # Converted to min

df['duration(hrs)'] = pd.to_timedelta(df['duration(min)'], unit = 'm') # Converted to h:m:s format
```

```python
df.columns
```

```python
df = df.drop('duration', axis = 1)
df = df.drop('listed_in', axis = 1)
df = df.drop('duration_numeric', axis = 1)
df = df.drop('durtion', axis = 1)


df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
df.columns
```

```python
df['Genre_1'] = df['Genre_1'].fillna('Unknown')
df['Genre_2'] = df['Genre_2'].fillna('Unknown')
df['Genre_3'] = df['Genre_3'].fillna('Unknown')
```

```python
df.to_csv('Cleaned_Netflix_movies.csv', index= False)
```

```python
df['Genre_1'].value_counts()

plt.figure(figsize =(30,10))
df['Genre_1'].value_counts().plot(kind= 'bar')
plt.title('Genre distribution ')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Genre_2'].value_counts()

plt.figure(figsize =(30,10))
df['Genre_2'].value_counts().plot(kind= 'bar', color = 'purple')
plt.title('Genre distribution ')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
