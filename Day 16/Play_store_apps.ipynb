```python
df = df.drop_duplicates()
```

```python
df.columns = ( df.columns.str.lower(). str.replace (' ', '_') .str.replace(r'[()]', '', regex = True))
```

```python
df.columns
```

```python
df['installs'] = df['installs'].str.replace('+', '', regex= False)
df['installs'] = df['installs'].str.replace(',', '', regex= False)

df['installs'] = pd.to_numeric(df['installs'],errors='coerce')
```

```python
df['price'] = df['price'].str.replace('$','',regex=False)

df['price'] = pd.to_numeric(df['price'],errors='coerce')
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mean())
```

```python
df['rating'] = pd.to_numeric(df['rating'],errors='coerce')
```

```python
df['type'] = df['type'].fillna('Unknown')
df['content_rating'] = df['content_rating'].fillna('Unknown')
df['current_ver'] = df['current_ver'].fillna('Unknown')
df['android_ver'] = df['android_ver'].fillna('Unknown')
```

```python
df['rating_class'] = df['rating'].apply(lambda x: 1 if x >= 4 else 0)
```

```python
# Features

X = df[['reviews','installs', 'price']]
y = df['rating_class']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(predictions)
```
