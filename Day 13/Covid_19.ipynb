```python
df['date_of_interest'] = pd.to_datetime(df['date_of_interest'])
```

```python
df = df.sort_values('date_of_interest')
```

```python
df.columns = ( df.columns.str.lower(). str.replace (' ', '_') .str.replace(r'[()]', '', regex = True))
```

```python
plt.figure
plt.plot(df['date_of_interest'],df['case_count'])
plt.title('Daily Covid 19 case')
plt.xlabel('Date')
plt.ylabel('No. of cases')
plt.show()
```

```python
plt.figure
plt.plot(df['date_of_interest'],df['death_count'])
plt.title('Daily Covid 19 case')
plt.xlabel('Date')
plt.ylabel('No. of deaths')
plt.show()
```
