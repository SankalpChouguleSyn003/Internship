```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Airbnb_Open_Data.csv')
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
df.columns
```

```python
df['name'] = df['name'].fillna('Unknown')
df['host_identity_verified'] = df['host_identity_verified'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
df['neighbourhood_group'] = df['neighbourhood_group'].fillna('Unknown')
df['neighbourhood'] = df['neighbourhood'].fillna('Unknown')
df['lat'] = df['lat'].fillna(df['lat'].median())
df['long'] = df['long'].fillna(df['long'].median())
df['country'] = df['country'].fillna('Unknown')
df['country_code'] = df['country_code'].fillna('Unknown')
df['instant_bookable'] = df['instant_bookable'].fillna('Unknown')
df['cancellation_policy'] = df['cancellation_policy'].fillna('Unknown')
df['construction_year'] = df['construction_year'].fillna('Unknown')
df['room_type'] = df['room_type'].fillna('Unknown')
df['room_type'] = df['room_type'].fillna('Unknown')
df['room_type'] = df['room_type'].fillna('Unknown')
df['price'] = df['price'].fillna('NA')
df['service_fee'] = df['service_fee'].fillna('Unknown')
df['minimum_nights'] = df['minimum_nights'].fillna(0)
df['number_of_reviews'] = df['number_of_reviews'].fillna(0)
df['last_review'] = df['last_review'].fillna(0)
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['review_rate_number'] = df['review_rate_number'].fillna(0)
df['calculated_host_listings_count'] = df['calculated_host_listings_count'].fillna(0)
df['availability_365'] = df['availability_365'].fillna(0)
df = df.drop(['license', 'house_rules'], axis=1)
```

```python
print(df.isnull().sum())
```

```python
# Price distribution

plt.hist(df['price'], bins=50)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df['number_of_reviews'], bins=40)
plt.title('Number of Reviews Distribution')
plt.xlabel('Reviews')
plt.ylabel('Count')
plt.show()
```

```python
plt.scatter(df['number_of_reviews'], df['price'])
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.title('Reviews vs Price')
plt.show()
```

```python
df.to_csv('cleaned_Airbnb_Open_Data.csv', index=False)
```
