```python
df = df.drop(columns = ['ehail_fee'])
```

```python
df['store_and_fwd_flag'] = df['store_and_fwd_flag'].fillna(df['store_and_fwd_flag'].mode())
df['payment_type'] = df['payment_type'].fillna(df['payment_type'].mode())
df['trip_type'] = df['trip_type'].fillna(df['trip_type'].mode())
```

```python
df['RatecodeID'] = df['RatecodeID'].fillna(df['RatecodeID'].median())
df['passenger_count'] = df['passenger_count'].fillna(df['passenger_count'].mode())
df['congestion_surcharge'] = df['congestion_surcharge'].fillna(df['congestion_surcharge'].mode())
```

```python
df['lpep_pickup_datetime'] = pd.to_datetime(df['lpep_pickup_datetime'],errors='coerce')
df['lpep_dropoff_datetime'] = pd.to_datetime(df['lpep_dropoff_datetime'],errors='coerce')
```

```python
df.to_csv('Cleaned_Green_Taxi_Trip.csv', index= True)
```

```python
# trip duration
df['trip_duration'] = (df['lpep_dropoff_datetime']- df['lpep_pickup_datetime']).dt.total_seconds() / 60

df['pickup_hour'] = df['lpep_pickup_datetime'].dt.hour
```

```python
# Trip duration distribution 

plt.figure()
plt.hist(df['trip_duration'], bins = 50)
plt.xlabel('Trip Duration (min)')
plt.ylabel('No of Trips')
plt.title('Trip Duration Distribution')
plt.show()
```

```python
# Fare vs Distance

plt.figure()
plt.scatter(df['trip_distance'], df['fare_amount'], alpha = 0.5)
plt.xlabel('Trip Distance')
plt.ylabel('Fare Amount')
plt.title('Fare vs Distance')
plt.show()
```

```python
peak_hours = df['pickup_hour'].value_counts().sort_index()

plt.figure()
plt.bar(peak_hours.index, peak_hours.values)
plt.xlabel("Hour of Day")
plt.ylabel("Number of Trips")
plt.title("Trips per Hour (Peak Hours)")

plt.show()
```
